```python
# /content/sample_data/4LDI.cif
# fopen = open("/content/sample_data/1ASY.pdb","r")

def pdb_read(file_path):
  total_AA = 0
  AA_number = None
  x_coord = []
  y_coord = []
  z_coord = []

  with open(file_path, 'r') as f:
        lines = f.readlines()

  for line in lines:
    if line.startswith("COMPND"):
        if "CHAIN" in line:
          print("Chains present:", line.strip())

  seq = str(input("Enter the chain_ID needed: ")).upper()
  for line in lines:
    if line.startswith("REMARK"):
      if "HIGH (ANGSTROMS)" in line:
          print(line.strip())
          # strip is used ,to remove any
          # whitespaces present between the lines.
          # total atoms
    if line.startswith("ATOM"):
      if line[21] == seq:
        # print(line)
        x_coord.append(float(line[31:38].strip()))
        y_coord.append(float(line[39:46].strip()))
        z_coord.append(float(line[47:54].strip()))
        if line[22:26].strip() != AA_number:
          total_AA +=1
          AA_number = line[22:26].strip()
  print(f"Total residues in chain {seq}: {total_AA}")

  print(f"The x_coordinate of COM is {sum(x_coord)/ len(x_coord)}")
  print(f"The y_coordinate of COM is {sum(y_coord)/ len(y_coord)}")
  print(f"The z_coordinate of COM is {sum(z_coord)/ len(z_coord)}")
```

```python
file_path = ("/content/sample_data/7YZW.pdb")
pdb_read(file_path)
```
