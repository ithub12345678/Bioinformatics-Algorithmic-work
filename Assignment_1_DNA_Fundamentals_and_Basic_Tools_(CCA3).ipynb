```python
"""
DNA Class for Basic Sequence Validation and Analysis
---------------------------------------------------
This class represents a DNA sequence and performs:
1. Sequence validation (ensuring only valid nucleotides are used)
2. Counting of nucleotides
3. Calculation of AT and GC content percentages
4. Calculation of individual nucleotide frequencies

Author: [Your Name]
Date: [Date]
"""
import re

class DNA:
    """A class to represent and analyze a DNA sequence."""

    def __init__(self):
        """
        Initialize the DNA object with:
        - A list of valid nucleotides (A, T, C, G)
        - A dictionary to count each nucleotide
        """
        self.valid_nucleotides = ["A", "T", "C", "G"]
        self.nucleotide_count = {"A": 0, "T": 0, "C": 0, "G": 0}
        self.seq_list = []
        self.codon_list = []


    def valid_dna(self, seq):
        """
        Validate and clean a DNA sequence.

        Args:
            seq (str): Input DNA sequence (can include lowercase or newlines).

        Returns:
            list: List of uppercase nucleotide characters (A, T, C, G).

        Raises:
            ValueError: If an invalid nucleotide is found in the sequence.
        """
        # Remove newline characters if present
        # and Convert to uppercase to ensure uniformity
        seq = seq.replace("\n", "", -1).upper()


        # Validate each character in the sequence


        for nucleotide in seq:
          validity =  bool(re.fullmatch(r'[ATGC]+',seq.upper()))
          if validity == False:
            if nucleotide not in self.valid_nucleotides:
                # Raise an error if an invalid character is found
                print(
                    f"Invalid nucleotide found in sequence: '{nucleotide}'")
                seq = seq.replace(nucleotide,"",-1)

                print(f"The replaced seq is : {seq}\n")
            else:
              continue

        for nucleotide in seq:
            self.seq_list.append(nucleotide)
        return self.seq_list

    def codon_seq(self,seq):

        for i in range(0,len(seq),3):
          self.codon_list.append(seq[i:i+3])
        return self.seq_list,self.codon_list


    def count(self, seq_list):
        """
        Return the total length of the DNA sequence.

        Args:
            seq_list (list): Validated list of nucleotides.

        Returns:
            str: Message showing the total length of the sequence.
        """
        return f"The length of the nucleotide sequence is: {len(seq_list)}"

    def map_count(self, seq_list):
        """
        Count the number of each nucleotide in the DNA sequence.

        Args:
            seq_list (list): List of validated nucleotides.

        Returns:
            dict: Dictionary with counts of A, T, C, and G.
        """
        # Count occurrences of each nucleotide
        self.nucleotide_count = {"A": 0, "T": 0, "C": 0, "G": 0}
        for nucleotide in seq_list:
            self.nucleotide_count[nucleotide] += 1

        return self.nucleotide_count

    def statistics(self, nucleotide_count):
        """
        Compute the AT and GC content of the DNA sequence.

        Args:
            nucleotide_count (dict): Dictionary of nucleotide counts.

        Returns:
            tuple: (AT_content_percentage, GC_content_percentage)
        """
        nucle_count = nucleotide_count

        # Calculate AT and GC content percentages
        a_t_content = (
            (nucle_count.get("A") + nucle_count.get("T")) /
            sum(nucle_count.values())
        ) * 100
        c_g_content = (
            (nucle_count.get("C") + nucle_count.get("G")) /
            sum(nucle_count.values())
        ) * 100

        return a_t_content, c_g_content

    def nucleotide_freq(self, nucleotide_count):
        """
        Calculate the percentage frequency of each nucleotide.

        Args:
            nucleotide_count (dict): Dictionary of nucleotide counts.

        Returns:
            tuple: (A_freq, T_freq, C_freq, G_freq)
        """
        nucle_count = self.nucleotide_count
        total = sum(nucle_count.values())

        # Calculate individual nucleotide frequencies in percentage
        a_freq = (nucle_count.get("A") / total) * 100
        t_freq = (nucle_count.get("T") / total) * 100
        c_freq = (nucle_count.get("C") / total) * 100
        g_freq = (nucle_count.get("G") / total) * 100

        return a_freq, t_freq, c_freq, g_freq

    def comparision(self, seq1, seq2):
        """
        Compare two DNA sequences based on AT and GC content.
        Returns the absolute difference between both sequences.
        """
        dna_seq1 = self.valid_dna(seq1)
        dna_seq2 = self.valid_dna(seq2)

        freq_seq1 = self.map_count(dna_seq1)
        freq_seq2 = self.map_count(dna_seq2)

        stats_seq1 = self.statistics(freq_seq1)
        stats_seq2 = self.statistics(freq_seq2)

        gc_diff = abs(stats_seq1[1] - stats_seq2[1])
        at_diff = abs(stats_seq1[0] - stats_seq2[0])

        return gc_diff, at_diff


# -----------------------
# Example Usage
# -----------------------
nucleotide_sequence = '''TAATAGTACAGACAGCbnvCCAATTATAAAATX'''

# Create DNA object
dna = DNA()

# # Step 1: Validate and clean the sequence
result = dna.valid_dna(nucleotide_sequence)
print("Validated codon seq List:", result)

# # The Detailed nucleotide_analysis_report -:
print("\nA comprehensive nucleotide analysis report -:\n")

# # Step 2: Get total count
count_total = dna.count(result)
print(count_total)

# # Step 3: Count nucleotide frequency
freq = dna.map_count(result)
print("\nNucleotide Counts:", freq)

# # Step 4: Calculate AT/GC content
statistics = dna.statistics(freq)
print("\nAT% and GC%:", statistics)

# # Step 5: Calculate individual nucleotide frequencies
nuc_freq = dna.nucleotide_freq(freq)
print("\nA%, T%, C%, G%:", nuc_freq)

# Step 6: Seq comparision
seq5 = "AAAAAAAAAAAAAAAATTTTTTTTTTTTTTT"
seq6 = "GGGGGGGGGGGGGGGGCCCCCCCCCCCCCCC"

compare = dna.comparision(seq5,seq6)
print("\nGC% and AT% content differences :",compare)
```

```python
# Building an effective transcription system.

import re


class DNA:
    """A class to represent and analyze DNA sequences."""

    def __init__(self):
        """
        Initialize the DNA object with:
        - A list of valid nucleotides (A, T, C, G)
        - A complement mapping dictionary for base pairing
        """
        self.valid_nucleotides = ["A", "T", "C", "G"]
        self.COMPLEMENT_MAP = {
            "A": "T", "T": "A", "G": "C", "C": "G",
            "R": "Y", "Y": "R", "S": "S", "W": "W",
            "K": "M", "M": "K", "B": "V", "V": "B",
            "D": "H", "H": "D", "N": "N"
        }

    def valid_dna(self, seq):
        """
        Validate and clean a DNA sequence.

        Args:
            seq (str): Input DNA sequence (may include lowercase letters or
                       invalid symbols).

        Returns:
            list: List of uppercase valid nucleotides.

        Notes:
            Invalid nucleotides are removed and reported.
        """
        # Remove newlines and convert to uppercase
        seq = seq.replace("\n", "").upper()

        # Identify invalid characters
        invalid_chars = re.findall(r"[^ATGC]", seq)
        if invalid_chars:
            print(f"Invalid characters found: {set(invalid_chars)}")
            # Remove invalid nucleotides
            seq = re.sub(r"[^ATGC]", "", seq)
            print(f"Cleaned sequence: {seq}\n")

        # Return as list of individual bases
        return list(seq)

    def codon_seq(self, seq):
        """
        Split a DNA sequence into codons (groups of 3 bases).

        Args:
            seq (list or str): Validated DNA sequence.

        Returns:
            list: List of codons.
        """
        if isinstance(seq, str):
            seq = list(seq)
        codons = [seq[i:i + 3] for i in range(0, len(seq), 3)]
        return codons

    def count(self, seq_list):
        """
        Return the total length of the DNA sequence.

        Args:
            seq_list (list): Validated list of nucleotides.

        Returns:
            str: Message showing the sequence length.
        """
        return f"The length of the nucleotide sequence is: {len(seq_list)}"

    def dna_to_rna(self, seq_list):
        """
        Transcribe a DNA template strand into mRNA and derive the coding DNA strand.

        Args:
            seq_list (list): Validated DNA sequence (template strand).

        Returns:
            tuple: (mRNA sequence, coding DNA strand)
        """
        # Transcribe DNA → RNA (template to mRNA)
        rna_seq_list = [
            "U" if n == "A" else
            "A" if n == "T" else
            "G" if n == "C" else
            "C"
            for n in seq_list
        ]

        # Coding strand has the same sequence as mRNA, except U → T
        coding_strand_dna = ["T" if n == "U" else n for n in rna_seq_list]

        return "".join(rna_seq_list), "".join(coding_strand_dna)

    def batch_processing(self, seqs_list):
        """
        Process multiple DNA sequences: clean, transcribe, and generate coding strands.

        Args:
            seqs_list (list): List of DNA sequences (strings).

        Returns:
            tuple: (cleaned sequences, mRNA sequences, coding DNA strands)
        """
        cleaned_seqs = []
        batch_rna_seqs = []
        batch_coding_strands = []

        for seq in seqs_list:
            valid_seq = self.valid_dna(seq)
            rna_seq, coding_strand = self.dna_to_rna(valid_seq)

            cleaned_seqs.append("".join(valid_seq))
            batch_rna_seqs.append(rna_seq)
            batch_coding_strands.append(coding_strand)

        return cleaned_seqs, batch_rna_seqs, batch_coding_strands

    def complement(self, seq):
        """
        Generate the complementary DNA strand.

        Args:
            seq (str): Input DNA sequence.

        Returns:
            list: Complementary DNA sequence.
        """
        seq = self.valid_dna(seq)
        return [self.COMPLEMENT_MAP[n] for n in seq]

    def rev_comp(self, seq):
        """
        Generate the reverse complement of a DNA strand.

        Args:
            seq (str): Input DNA sequence.

        Returns:
            tuple: (reverse complement 5'→3', complement 3'→5')
        """
        complemented_seq = self.complement(seq)
        reverse_complemented_seq = complemented_seq[::-1]

        return reverse_complemented_seq, complemented_seq
```

```python
"""
DNA Sequence Analysis and Transcription System
----------------------------------------------
Performs:
1. Sequence validation and cleaning
2. Nucleotide counting and composition analysis
3. AT/GC content and frequency statistics
4. DNA→mRNA transcription and coding strand synthesis
5. Complement and reverse complement generation
6. Batch processing of multiple sequences

Author : Anshul Laxane
Date   : 24 Oct 2025
"""

import re
from collections import Counter


class DNA:
    """A unified class for DNA sequence analysis and transcription."""

    def __init__(self):
        """Initialize valid nucleotides and complement map."""
        self.valid_nucleotides = {"A", "T", "C", "G"}
        self.complement_map = str.maketrans(
            "ATGCRYSWKMBDHVNatgcryswkmbdhvn",
            "TACGYRSWMKVHDBNtacgyrswmkvhdbn"
        )

    # ------------------ Core Validation ------------------ #
    def clean(self, seq):
        """Validate and clean DNA sequence, removing invalid characters."""
        seq = re.sub(r"[^ATGCatgc]", "", seq).upper()
        return seq

    # ------------------ Sequence Analysis ------------------ #
    def count_bases(self, seq):
        """Return counts of each nucleotide."""
        seq = self.clean(seq)
        return dict(Counter(seq))

    def length(self, seq):
        """Return length of the sequence."""
        return len(self.clean(seq))

    def gc_at_content(self, seq):
        """Calculate AT% and GC% composition."""
        seq = self.clean(seq)
        total = len(seq) or 1
        gc = (seq.count("G") + seq.count("C")) / total * 100
        at = (seq.count("A") + seq.count("T")) / total * 100
        return round(at, 2), round(gc, 2)

    def freq_percent(self, seq):
        """Return percentage frequency of each nucleotide."""
        seq = self.clean(seq)
        total = len(seq) or 1
        counts = Counter(seq)
        return {base: round((count / total) * 100, 2) for base, count in counts.items()}

    # ------------------ Transcription System ------------------ #
    def dna_to_rna(self, seq):
        """Transcribe DNA template strand to mRNA and derive coding strand."""
        seq = self.clean(seq)
        transcription = {"A": "U", "T": "A", "C": "G", "G": "C"}
        rna = "".join(transcription.get(base, "") for base in seq)
        coding_dna = rna.replace("U", "T")
        return rna, coding_dna

    def codons(self, seq):
        """Split DNA into codons (groups of 3 bases)."""
        seq = self.clean(seq)
        return [seq[i:i + 3] for i in range(0, len(seq), 3)]

    # ------------------ Sequence Comparison ------------------ #
    def compare(self, seq1, seq2):
        """Compare two sequences by AT/GC content difference."""
        at1, gc1 = self.gc_at_content(seq1)
        at2, gc2 = self.gc_at_content(seq2)
        return abs(gc1 - gc2), abs(at1 - at2)

    # ------------------ DNA Strand Operations ------------------ #
    def complement(self, seq):
        """Return complementary DNA strand."""
        return self.clean(seq).translate(self.complement_map)

    def reverse_complement(self, seq):
        """Return reverse complement of DNA strand."""
        comp = self.complement(seq)
        return comp[::-1], comp

    # ------------------ Batch and Merge Utilities ------------------ #
    def batch_process(self, sequences):
        """Batch clean and transcribe multiple DNA sequences."""
        cleaned, rnas, codings = [], [], []
        for seq in sequences:
            seq = self.clean(seq)
            rna, coding = self.dna_to_rna(seq)
            cleaned.append(seq)
            rnas.append(rna)
            codings.append(coding)
        return cleaned, rnas, codings

    def merge_fragments(self, fragments):
        """Merge multiple DNA fragments into one continuous sequence."""
        return "".join(frag.upper() for frag in fragments)


# ----------------------- Example Usage ----------------------- #
if __name__ == "__main__":
    dna = DNA()

    seq = "TAATAGTACAGACAG*$CbnvCCAATTATAAAATX"
    print("Cleaned Sequence:", dna.clean(seq))
    print("Base Count:", dna.count_bases(seq))
    print("Sequence Length:", dna.length(seq))
    print("AT% / GC%:", dna.gc_at_content(seq))
    print("Frequency (%):", dna.freq_percent(seq))
    print("Codons:", dna.codons(seq))

    # Transcription
    rna, coding = dna.dna_to_rna(seq)
    print("\nTranscription → mRNA:", rna)
    print("Coding Strand DNA:", coding)

    # Comparison
    seq1, seq2 = "AAAAATTTTT", "GGGGCCCCCC"
    print("\nGC% / AT% Difference:", dna.compare(seq1, seq2))

    # Complement and Reverse Complement
    print("\nComplement:", dna.complement(seq1))
    print("Reverse Complement:", dna.reverse_complement(seq1)[0])

    # Batch Processing
    seqs = ["atgccta", "ttgcag", "cgatat"]
    print("\nBatch Processing:", dna.batch_process(seqs))

    # Merge Fragments
    print("Merged Sequence:", dna.merge_fragments(["atcg", "ccgta", "agcta"]))
```
